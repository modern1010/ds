```python
import numpy as np
import matplotlib.pyplot as plt

# Install necessary packages
!pip install numpy matplotlib scikit-learn

# Import required libraries after installation
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Data
x = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)  # Reshape to a column vector
y = np.array([7, 14, 15, 18, 19, 21, 26, 23])

# Simple Linear Regression
model = LinearRegression()
model.fit(x, y)

# Coefficients
b0 = model.intercept_
b1 = model.coef_[0]

# Predictions
y_pred = model.predict(x)

# Model Evaluation
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Plotting
plt.scatter(x, y, label='Actual Data')
plt.plot(x, y_pred, color='red', label=f'Regression Line: y = {b0:.2f} + {b1:.2f}x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Simple Linear Regression')
plt.show()

# Display Results
print(f'Intercept (b0): {b0:.2f}')
print(f'Slope (b1): {b1:.2f}')
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R-squared (R2): {r2:.2f}')
```
